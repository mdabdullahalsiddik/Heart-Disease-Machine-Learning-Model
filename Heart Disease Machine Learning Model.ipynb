# Training & Testing Data Model

```python
# input train_test_split 
from sklearn.model_selection import train_test_split 
```

```python
# Split the data into training and testing sets with a ratio of 70:30
xtrain,xtest,ytrain,ytest = train_test_split(x_n,y_n,test_size=0.30,random_state=42)
```

# LogisticRegression

```python
# input LogisticRegression 
from sklearn.linear_model import LogisticRegression  
```

```python
# createing LogisticRegression object
logr = LogisticRegression() 
```

```python
# Train the model on the training data
logr.fit(xtrain,ytrain)
```
